```python
def spectrum_in_bins(df,m,M,bin_size):
    # Now, let's define the mz ranges, and the label associated to each of them (the mean of the limiting values of each bin)
    range_min = []; range_max = []; range_label = [];
    for mz in range(m,M,bin_size):
        range_min.append(mz)
        range_max.append(mz+bin_size)
        range_label.append(np.mean([range_min[-1],range_max[-1]]).astype(int))


    N = len(df)  # number of samples
    L = len(range_min)  # length of new spectrum (number of bins)
    all_data = np.zeros((N,L))
    for idx in range(N): 
        intensity = df[['intensity']].iloc[idx].values[0]
        mzcoord   = df[['coord_mz']].iloc[idx].values[0]
        idx_data_in_bins = np.zeros((1,L))
        for i,mz in enumerate(range_min):
            intensity_range = intensity[(mzcoord > mz) & (mzcoord < (mz+bin_size))]
            if len(intensity_range) > 0 :
                idx_data_in_bins[0,i] = np.max(intensity_range)
            else: # if those mz coordinates are not in that spectrum
                idx_data_in_bins[0,i] = 0   

        # Normalize the amplitude of the spectrum
        idx_data_in_bins[0,:] = idx_data_in_bins[0,:] / np.max(idx_data_in_bins[0,:])
        
        
        all_data[idx,:] = idx_data_in_bins
    new_df = pd.DataFrame(data=all_data, columns = range_label, index = df.index)
    return new_df
```

### Data split
We assume that we are going to be performing cross-validation, so no validation set is needed.

```python
from sklearn.model_selection import train_test_split

# Extract data (spectra) and targets of the df_train set
data = df_train.iloc[:,-2:]
targets = df_train.iloc[:,1:-2]

m = 2000; M = 20000; 
bin_size = 50;

# Then, split into a train and test_train set
data_train, data_test_train, targets_train, targets_test_train = train_test_split(data, targets, test_size=0.2, random_state=42) # split the data
print('Training samples: '+str(len(data_train))+' and test_train samples: ' + str(len(data_test_train)) )

# apply the bins to all spectra, so that our feature space becomes the same for all samples
spectrum_train = spectrum_in_bins(data_train,m,M,bin_size)
spectrum_test_train = spectrum_in_bins(data_test_train,m,M,bin_size)
# these spectrum_... are our X for training
```

# RF
As a first trial, let's try a RF

```python
from sklearn.ensemble import RandomForestClassifier 
RF = RandomForestClassifier(n_estimators = 70, max_depth = 3)
RF.fit(spectrum_train.values,targets_train[['OXACILINA']].to_numpy().reshape(-1,))

acc_train = RF.score(spectrum_train,targets_train[['OXACILINA']].to_numpy().reshape(-1,))
acc_test = RF.score(spectrum_test_train,targets_test_train[['OXACILINA']].to_numpy().reshape(-1,))
print('Train accuracy is %.4f and test accuracy is %.4f' %(acc_train,acc_test))
```

We will be evaluated with the AUC (area under curve), where the curve is the ROC (receiver operating characteristics). Let's play a bit with those functions:

```python
pred_train = RF.predict_proba(spectrum_train)
pred_test = RF.predict_proba(spectrum_test_train)[:,-1]
```

```python
Y_test = targets_test_train[['OXACILINA']].to_numpy().reshape(-1,)
print(Y_test.shape)
print(pred_test.shape)
print(pred_test)
```

```python
from sklearn.metrics import roc_curve, auc, roc_auc_score
# roc_auc_score method
auc_score = roc_auc_score(Y_test,pred_test)
print('ROC_AUC_score method: \t\t',auc_score)

# roc_curve + auc methods (same, but also enables plotting)
fpr, tpr, _ = roc_curve(Y_test,pred_test)
roc_auc = auc(fpr, tpr)
print ('ROC_curve + ROC_AUC method: \t',roc_auc )

plt.plot(fpr, tpr, label='ROC curve of class')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC curve per class')
plt.legend(loc="lower right")
plt.show()
```

# Function to try different classifiers
The goal is a function to which we input a classifier "clf", a set of paramenters "params", and it performs a cross-validated search for the nine different antibiotics, providing back the AUC, accuracy, and the trained classifiers for all of them.

```python
# First, let's make it with code, later put into a function
clf = RandomForestClassifier()
params = {'n_estimators':[10,20,30,40,50,70,100,150,200],'max_depth':np.arange(1,20)}
from sklearn.model_selection import GridSearchCV

# first, for just one of the categories (the first, oxacilina)
antibiotic_idx = 0
antibiotic_targets_train = targets_train.iloc[:,antibiotic_idx].to_numpy().reshape(-1,)
antibiotic_targets_test_train = targets_test_train.iloc[:,antibiotic_idx].to_numpy().reshape(-1,)

grid = GridSearchCV(clf,param_grid=params, cv=5, iid=False)
grid.fit(spectrum_train, antibiotic_targets_train)

print('The best parameters are: ',grid.best_params_)
best_clf = grid.best_estimator_
# later, save in a list of best estimators that will be returned by the function

acc_train = best_clf.score(spectrum_train, antibiotic_targets_train)
acc_test = best_clf.score(spectrum_test_train, antibiotic_targets_test_train)
print('Train accuracy: ',acc_train,' and test_train accuracy: ',acc_test)

pred_train = best_clf.predict_proba(spectrum_train)[:,-1] # only take last column, the prob of Y = +1
pred_test = best_clf.predict_proba(spectrum_test_train)[:,-1]
auc_score_train = roc_auc_score(antibiotic_targets_train, pred_train)
auc_score_test = roc_auc_score(antibiotic_targets_test_train, pred_test)
print('Train AUC: ',auc_score_train,' and Test_train AUC: ',auc_score_test)

```
