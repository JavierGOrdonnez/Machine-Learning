```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
path = "D:/GitHub/Machine-Learning/Kaggle/"

# read the gp file
df_submission = pd.read_csv(path+'gps_lowarea.csv') 
categories = df_submission.columns[1:]
for c, cat in enumerate(categories): 
    plt.hist(df_submission[cat],20)
    plt.title(cat)
    plt.show()

# # Save the dataframe with the predicted outputs
# df_submission = df_submission.set_index('ID')
# df_submission.to_csv(path_results + file_name + '.csv')
# print('DONE!')
```

```python
new_df_submission = df_submission.copy(deep=True)
for c, cat in enumerate(categories): 
    vector = new_df_submission[[cat]].to_numpy()
    vector[vector>1]=1
    vector[vector<0]=0
    new_df_submission[[cat]] = vector
```

```python
categories = new_df_submission.columns[1:]
for c, cat in enumerate(categories): 
    plt.hist(new_df_submission[cat],20)
    plt.title(cat)
    plt.show()
```
